```python
import matplotlib.pyplot as plt
from glob import glob
import os
import numpy as np
```

```python
PATH = 'C:/Users/user/Desktop/Child Skin Disease'
DATA_PATH = os.path.join(PATH, 'Total_Dataset')
```

```python
label_to_index = {}

for i in range(12):
    for dic in os.listdir(os.path.join(DATA_PATH, 'H' + str(i))):
        label_to_index[dic] = 0
    
label_to_index
```

```python
def get_data(num): 
    
    data = {}
    
    for i in range(12):
        dataset = os.path.join(DATA_PATH, f'H{i}')
        files = os.listdir(dataset)
        
        for f in files: 
            imgs = glob(os.path.join(dataset, f) + '/*.jpg')
            print(len(imgs))
            if len(imgs) > num:
                data[f] = len(imgs)
            # else:
                # pass

    return data 
        
    
```

```python
# for idx, val in label_to_index:
#     print(val)
data = get_data(500) 
data
```

```python
#  
names = list(label_to_index.keys())
values = list(label_to_index.values())

plt.figure(figsize=(30, 25))
plt.barh(range(len(label_to_index)), values, tick_label=names)
plt.show()   
```

```python
# Train
train_dict = {}

for i in range(6):
    dataset = os.path.join(DATA_PATH, f'H{i}')
    files = os.listdir(dataset)
    
    for f in files: 
        imgs = glob(os.path.join(dataset, f) + '/*.jpg')
        # print(os.path.join(dataset, f))
        # print(len(imgs))
        # print(imgs)
        # print(f)
        # label_to_index[f] = label_to_index[f] + len(imgs)
        
        if f in train_dict:
            train_dict[f] += len(imgs)
        else:
            train_dict[f] = len(imgs)
            

train_names = list(train_dict.keys())
train_values = list(train_dict.values())

plt.figure(figsize=(30, 25))
plt.barh(range(len(train_dict)), train_values, tick_label=train_names)
plt.show()         
    
```

```python
# test
test_dict = {}

for i in range(7, 10):
    dataset = os.path.join(DATA_PATH, f'H{i}')
    files = os.listdir(dataset)
    
    for f in files: 
        imgs = glob(os.path.join(dataset, f) + '/*.jpg')
        # print(os.path.join(dataset, f))
        # print(len(imgs))
        # print(imgs)
        # print(f)
        # label_to_index[f] = label_to_index[f] + len(imgs)
        
        if f in test_dict:
            test_dict[f] += len(imgs)
        else:
            test_dict[f] = len(imgs)
            

test_names = list(test_dict.keys())
test_values = list(test_dict.values())

plt.figure(figsize=(30, 25))
plt.barh(range(len(test_dict)), test_values, tick_label=test_names)
plt.show()      
```

```python
bar_width = 0.5
plt.figure(figsize=(20, 50))
plt.barh(np.arange(len(train_dict)), train_values, bar_width, tick_label=train_names, color='r')
plt.barh(np.arange(len(test_dict)) + bar_width , test_values, bar_width, tick_label=test_names)
plt.legend(('Train', 'Test'))
plt.show()  
```
